```coconut
sample_input="""
16
10
15
5
1
11
7
19
6
12
4"""

adapter_list = sample_input.splitlines() |> filter$(None) |> map$(int) |> list
adapter_list.sort()
built_in_adapter = adapter_list[-1] + 3
adapter_list.append(built_in_adapter)

def count_distribution(adapter_list):
    dists = {}
    last = 0
    it = iter(adapter_list)
    for element in it:
        diff = element - last
        if diff > 3 or diff < 1:
            print("invalid difference:", last, element, diff)
            break
        if diff in dists:
            dists[diff] += 1
        else:
            dists[diff] = 1
        last = element
    return dists

result = count_distribution(adapter_list)
print(repr(result))
assert(result == {1: 7, 3: 5})

sample_input2="""
28
33
18
42
31
14
46
20
48
47
24
23
49
45
19
38
39
11
1
32
25
35
8
17
7
9
4
2
34
10
3"""

adapter_list = sample_input2.splitlines() |> filter$(None) |> map$(int) |> list
adapter_list.sort()
built_in_adapter = adapter_list[-1] + 3
adapter_list.append(built_in_adapter)

result = count_distribution(adapter_list)
print(repr(result))
assert(result == {1: 22, 3: 10})


puzzle_input="""
160
34
123
159
148
93
165
56
179
103
171
44
110
170
147
98
25
37
137
71
5
6
121
28
19
134
18
7
66
90
88
181
89
41
156
46
8
61
124
9
161
72
13
172
111
59
105
51
109
27
152
117
52
68
95
164
116
75
78
180
81
47
104
12
133
175
16
149
135
99
112
38
67
53
153
2
136
113
17
145
106
31
45
169
146
168
26
36
118
62
65
142
130
1
140
84
94
141
122
22
48
102
60
178
127
73
74
87
182
35"""

adapter_list = puzzle_input.splitlines() |> filter$(None) |> map$(int) |> list
adapter_list.sort()
built_in_adapter = adapter_list[-1] + 3
adapter_list.append(built_in_adapter)

result = count_distribution(adapter_list)
print("Answer:", repr(result[1] * result[3]))
```
