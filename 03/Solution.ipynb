```coconut
def check_slope(tree_map, slope):
    position = (0, 0)
    height = len(tree_map)
    width = len(tree_map[0])
    
    trees = 0
    
    while position[0] < height:
        if tree_map[position[0]][position[1] % width] == '#':
            trees += 1
        position = (position[0] + slope[0], position[1] + slope[1])
    
    return trees

slopes = [
    (1, 1),
    (1, 3),
    (1, 5),
    (1, 7),
    (2, 1),
]

trees_product = [check_slope(sample_input, slope) for slope in slopes] |> reduce$((*), ?)
print(trees_product)
assert(trees_product == 336)

trees_product = [check_slope(puzzle_input, slope) for slope in slopes] |> reduce$((*), ?)
print("Answer:", trees_product)
```
